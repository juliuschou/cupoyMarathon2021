* 在精簡深度學習的方式上 : 卷積類神經 (CNN) 採用像素遠近，而遞歸類神經 (RNN) 採用著則是時間遠近
* 那麼，既然有著類似的設計精神，兩者是否有可能互換應用呢?

CNN 與 RNN分別適用於不同的目的，故不存在是否可以替代使用的議題，而是根據我們要解決的問題，選擇適當的model。

CNN 最常應用於電腦視覺的識別圖片中物件，例如: 識別動物種類(貓、狗)，或是水果種類等，如下圖所示:




RNN 適用於有順序性的資料，比較像是特徵的萃取，例如以下文字:

President **Roosevelt** was one of the most influential presidents in American history. However, **Roosevelt** Street in Manhattan was not named after him.

第一個句子應被Roosevelt應被標識為人物，然而在第二個句子應被標識為街道名稱或是位置。進行這樣的判斷需要依賴Roosevelt前面的 President 以及後面的 Street。
